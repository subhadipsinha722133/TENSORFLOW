```python
x=a.iloc[:,:-1]
y=a.iloc[:,-1]
```

```python
x
```

```python
from imblearn.over_sampling import RandomOverSampler

ros=RandomOverSampler()
x,y=ros.fit_resample(x,y)
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
from keras import Sequential
from keras.layers import Dense
from keras.callbacks import EarlyStopping
```

```python
model=Sequential()

model.add(Dense(20,input_dim=12,activation="sigmoid"))
model.add(Dense(900,activation="relu"))
model.add(Dense(800,activation="softmax"))
model.add(Dense(2000,activation="relu"))
model.add(Dense(600,activation="tanh"))
model.add(Dense(200,activation="sigmoid"))
model.add(Dense(1,activation="sigmoid"))


```

```python
model.summary()
```

```python
call=EarlyStopping(monitor= "val_loss",
    min_delta= 0,
    patience= 20,
    verbose= 0.2,
    mode= "auto",
    baseline= None,
    # restore_best_weight= False,
    start_from_epoch= 0)
```

```python

model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
h=model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=800,callbacks=call)
```

```python
import matplotlib.pyplot as plt

plt.plot(h.history["loss"])
plt.plot(h.history["val_loss"])

plt.show()
```
