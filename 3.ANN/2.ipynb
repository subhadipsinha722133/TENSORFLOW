```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
st=StandardScaler()
x_train=st.fit_transform(x_train)
x_test=st.transform(x_test)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.pairplot(a)
```

```python
"""apply maecine leaning algo================="""

from sklearn.linear_model import LinearRegression
li=LinearRegression()
li.fit(x_train,y_train)
print(li.score(x_test,y_test)*100)
```

```python
'''apply deep learning algo===================='''


from keras import Sequential
from keras.layers import Dense
```

```python
model=Sequential()

model.add(Dense(20,input_dim=17,activation="linear"))
model.add(Dense(2000,activation="linear"))
# model.add(Dense(3000,activation="relu"))
model.add(Dense(300,activation="linear"))
model.add(Dense(300,activation="linear"))
# model.add(Dense(300,activation="relu"))
model.add(Dense(300,activation="linear"))
model.add(Dense(800,activation="linear"))

model.add(Dense(1,activation="linear"))

model.compile(optimizer="adam",loss="mse")
model.summary()
```

```python
model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=300,validation_split=0.02)
```

```python
pred=model.predict(x_test)
pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,pred)*100
```
