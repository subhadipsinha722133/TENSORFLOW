```python
x=a.iloc[:,:-1]
y=a.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
from keras.layers import Dense
from keras import Sequential
from keras.optimizers import Adam
```

```python
modal=Sequential()


modal.add(Dense(40,input_dim=33,activation="linear"))
modal.add(Dense(50,activation="linear"))
# modal.add(Dense(60,activation="linear"))
modal.add(Dense(70,activation="linear"))
# modal.add(Dense(80,activation="linear"))
modal.add(Dense(1,activation="linear"))

modal.compile(Adam(learning_rate=0.2),loss="mse")
```

```python
modal.summary()
```

```python
modal.fit(x_train,y_train,epochs=3,validation_data=(x_test,y_test))
```

```python
pred=modal.predict(x_test)
pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,pred)
```
